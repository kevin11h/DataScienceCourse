## [Scenario 1:] Begin by providing an estimate for the fraction of students who played a video game in the week prior to the survey. Provide an interval estimate as well as a point estimate for this proportion.

```python
%%R

playweekbefore <- videodata$time!= 0.0
print(sum(playweekbefore))
fractionplayweekbefore <- sum(playweekbefore) / 91
print('Point Estimate for π:')
print(fractionplayweekbefore)
SE = sqrt(fractionplayweekbefore*(1-fractionplayweekbefore)/(91 - 1)) * sqrt((314-91)/314) 
print('SE value: ')
print(SE)
E = qnorm(0.975)*SE
print('Interval Estimate: ')
fractionplayweekbefore + c(-E, E)
```

```python
%%R
time = videodata$time
bootstrappop = as.vector(sapply(time, rep.int, times=4))

# Create a huge matrix with 10000 rows (one for each bootstrap sample) and 91 columns (one for each sampled value, to match the original sample size)
B = 1e4
n = nrow(videodata)
boot.samples = matrix(sample(bootstrappop, size = B * n, replace = TRUE), B, n)
# The three arguments to apply() are the object on which parts will applied the function, the number 1 to indicate the function should be applied to each row (use 2 for #columns), and the function name mean
boot.statistics = apply(boot.samples, 1, mean)
# Graph this with a density plot overlaying a histogram for something different

ggplot(data.frame(meanTime = boot.statistics),aes(x=meanTime)) +
  geom_histogram(binwidth=0.25,aes(y=..density..)) +
  geom_density(color="red")
```

### Using PCA technique

```python
def one_of_k_encode_where(df):
    codetable = dict(zip(range(1,7), ['arcade',
                                      'home_system',
                                      'home_computer',
                                      'arcade_and_either_home_computer_of_system',
                                      'home_computer_and_system',
                                      'all_three']))

    for (numeric_categorical_label, where) in codetable.iteritems():
        df['yes_' + where] = map(lambda w : 1 if w == numeric_categorical_label else 0, df['where'])
        df['not_' + where] = 1 - df['yes_' + where]
        
    return df
        
def one_of_k_encode_how_often(df):
    codetable = dict(zip(range(1,5), ['daily', 'weekly', 'monthly', 'semesterly']))
    
    for (numeric_categorical_label, when) in codetable.iteritems():
        df['yes_' + when] = map(lambda w : 1 if w == numeric_categorical_label else 0, df['freq'])
        df['not_' + when] = 1 - df['yes_' + when]
        
    return df
```

```python
VDF = one_of_k_encode_where(one_of_k_encode_how_often(VDF))
VDF = VDF[list(set(VDF.columns) - set(['where', 'freq']))]
```

```python
NEVER_PLAYED = 1
NON_RESPONSE = 99

has_played = VDF[(VDF.like != NEVER_PLAYED)]
rows_with_non_response = has_played.apply(lambda row : any([v == NON_RESPONSE for v in row]), axis=1)
has_played_complete_response_df = has_played[~rows_with_non_response]
has_played_complete_response_df = has_played_complete_response_df[sorted(has_played_complete_response_df.columns)]
```

```python
from sklearn.decomposition import PCA

corr_training_matrix = np.corrcoef(has_played_complete_response_df.T.astype(np.int64))
pca_fit_model = PCA().fit(corr_training_matrix)
```

```python
plt.figure(figsize=(12,8))
plt.bar(np.arange(33)+1, pca_fit_model.explained_variance_ratio_.cumsum(), width=.99)
plt.ylim((0,1))
plt.xlim((1,33.9))
plt.xlabel('PC by decreasing eigenvalue')
plt.ylabel('Cumulative Explained Variance')
plt.title('Scree Plot')
```

```python
loadings_df = DataFrame(pca_fit_model.components_.T,
                        index = has_played_complete_response_df.columns,
                        columns = ['PC%d'%(i+1) for i in range(33)])
loadings_df
```

```python
fig, axes = plt.subplots(nrows=1,ncols=6, figsize=(20,24), sharey=True)

for (ith, ax) in enumerate(axes):    
    ax.barh(np.arange(33)+1, pca_fit_model.components_[ith])
    ax.set_title('Component PC%d' % (ith+1))
    ax.set_xlabel('Loading Value')
    ax.set_xlim((-1,1))
    ax.set_ylim(1,33.9)
    ax.set_yticks(np.arange(33)+1.5)
    ax.set_yticklabels(has_played_complete_response_df.columns)
```

Since the top 6 Principal Components Explain ~80% of the Variance, we'll sort by the loadings dataframe by magnitude:

```python
N = 20

for i in range(6):
    pc_column_label_name = 'PC%d'%(i+1)
    loadings_df[pc_column_label_name + '_magnitude'] = loadings_df[pc_column_label_name].apply(abs)
    print 'Top %d Important Factors for' % N, pc_column_label_name
    print loadings_df.sort(pc_column_label_name + '_magnitude', ascending=False)[pc_column_label_name + '_magnitude'][:N]
    print "\n\n"
```
